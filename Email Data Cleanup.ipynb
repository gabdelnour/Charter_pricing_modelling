```python
import numpy as np
import pandas as pd
```

```python
email_df = pd.read_csv('data-2024-07-16.csv', sep=';')
```

```python
email_df.head()
```

```python
columns_to_drop = ['Message ID', 'Received At', 'Processed At', 'Date', 'Source', 'Aircraft', 'Source Suggested Price', 'Source Suggested Price.1', 'Source Suggested Price.2', 'Source Suggested Price.3', 'Source Suggested Price.4', 'Phone', 'Mobile', 'Email Subject']
filtered_email_df = email_df.drop(columns=columns_to_drop)
filtered_email_df.head()
```

```python
filtered_email_df.info()
```

```python
import re
```

```python
# Define a function to extract Name, Email, and Phone Number with the correct format
def extract_customer_info_corrected(customer_info):
    if isinstance(customer_info, str):
        name_match = re.match(r"([^()]+)", customer_info)
        email_match = re.search(r"\(([^,]+)", customer_info)
        phone_match = re.search(r",\s*([^()]+)\)", customer_info)
        
        name = name_match.group(1).strip() if name_match else None
        email = email_match.group(1).strip() if email_match else None
        phone = phone_match.group(1).strip() if phone_match else None
        
        return pd.Series([name, email, phone])
    else:
        return pd.Series([None, None, None])
        
customer_info_corrected_df = filtered_email_df['Customer'].apply(extract_customer_info_corrected)
customer_info_corrected_df.columns = ['Name', 'Email', 'Phone Number']

df_corrected = pd.concat([filtered_email_df, customer_info_corrected_df], axis=1)
```

```python
df_corrected
```

```python
# Define a function to split the Trip Start Date/time into Date and Time
def split_trip_date_time(trip_start):
    if isinstance(trip_start, str):
        date_match = re.search(r"(\d{2}/\d{2}/\d{4})", trip_start)
        time_match = re.search(r"@ ([\d:apm]+)", trip_start)
        
        date = date_match.group(1).strip() if date_match else None
        time = time_match.group(1).strip() if time_match else None
        
        return pd.Series([date, time])
    else:
        return pd.Series([None, None])

# Apply the function to the 'Trip Start Date/time' column
trip_start_date_time_df = df_corrected['Trip Start Date/time'].apply(split_trip_date_time)
trip_start_date_time_df.columns = ['Trip Start Date', 'Trip Start Time']

# Apply the function to the 'Trip End Date/time' column
trip_end_date_time_df = df_corrected['Trip End Date/time'].apply(split_trip_date_time)
trip_end_date_time_df.columns = ['Trip End Date', 'Trip End Time']

# Add the new columns to the original dataframe
df = pd.concat([df_corrected, trip_start_date_time_df, trip_end_date_time_df], axis=1)
```

```python
df
```

```python
df = df.drop(columns=['Customer', 'Trip Start Date/time', 'Trip End Date/time'])
```
