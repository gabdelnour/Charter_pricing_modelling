```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('us-airports.csv')
```

```python
df.head(50)
```

```python
df.shape
```

```python
df['type'].unique()
```

```python
airport_types = ['large_airport', 'medium_airport',
       'small_airport']

airport_df = df[df['type'].isin(airport_types)]

airport_df.head(50)
```

```python
airport_df.shape
```

```python
columns_to_drop = ['continent', 'country_name', 'iso_country', 'iso_region', 'municipality', 'scheduled_service', 'gps_code', 'iata_code', 'local_code', 'home_link', 'wikipedia_link', 'keywords', 'score', 'last_updated']
```

```python
airport_df = airport_df.drop(columns=columns_to_drop)
airport_df.head()
```
